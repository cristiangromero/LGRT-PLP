```python
class Producto:
    def __init__(self, nombre, precio, stock):
        # Programación defensiva: validación en el origen de los datos
        if precio < 0:
            raise ValueError(f"Error: El precio de '{nombre}' no puede ser negativo.")
        if stock < 0:
            raise ValueError(f"Error: El stock de '{nombre}' no puede ser negativo.")
        self.nombre = nombre
        self.precio = precio
        self.stock = stock

class Inventario:
    def __init__(self):
        self.productos = {}

    def agregar_producto(self, producto):
        if isinstance(producto, Producto):
            self.productos[producto.nombre] = producto

    def vender(self, nombre_producto, cantidad):
        try:
            # 1. Verificación explícita de existencia (Guard Clause)
            if nombre_producto not in self.productos:
                raise KeyError(f"El producto '{nombre_producto}' no existe en el catálogo.")

            # 2. Validación de cantidad positiva
            if cantidad <= 0:
                print("Error: La cantidad a vender debe ser mayor a cero.")
                return 0

            producto = self.productos[nombre_producto]

            # 3. Validación de stock suficiente
            if cantidad > producto.stock:
                print(f"Error: Stock insuficiente. Disponible: {producto.stock}")
                return 0

            # Operación segura
            producto.stock -= cantidad
            print(f"Venta exitosa: {cantidad} x {nombre_producto}")
            return producto.precio * cantidad

        except KeyError as e:
            print(f"Error de Inventario: {e}")
            return 0

    def obtener_valor_total(self):
        # El uso de variables locales aquí ocurre en el Stack [9]
        total = 0
        for p in self.productos.values():
            total += p.precio * p.stock
        return total
```

### Demostración del uso de las clases `Producto` e `Inventario`

```python
print('--- Creando productos ---')
producto1 = Producto('Manzanas', 1.5, 100)
producto2 = Producto('Naranjas', 2.0, 50)
producto3 = Producto('Peras', 3.0, 75)

print(f"Producto 1: {producto1.nombre}, Precio: {producto1.precio}, Stock: {producto1.stock}")
print(f"Producto 2: {producto2.nombre}, Precio: {producto2.precio}, Stock: {producto2.stock}")
print(f"Producto 3: {producto3.nombre}, Precio: {producto3.precio}, Stock: {producto3.stock}")
```

### Creación de un inventario y adición de productos

```python
print('\n--- Inicializando Inventario y agregando productos ---')
inventario = Inventario()
inventario.agregar_producto(producto1)
inventario.agregar_producto(producto2)
inventario.agregar_producto(producto3)

print(f"Productos en inventario: {list(inventario.productos.keys())}")
print(f"Valor total inicial del inventario: {inventario.obtener_valor_total():.2f}")
```

### Realizando ventas (casos exitosos y con errores)

```python
print('\n--- Realizando ventas ---')

# Venta exitosa
print('\n--- Venta exitosa de Manzanas ---')
monto_venta1 = inventario.vender('Manzanas', 10)
print(f"Monto de la venta 1: {monto_venta1:.2f}")
print(f"Stock actual de Manzanas: {inventario.productos['Manzanas'].stock}")

# Intento de venta de producto inexistente
print('\n--- Intento de venta de un producto que no existe ---')
monto_venta2 = inventario.vender('Uvas', 5)
print(f"Monto de la venta 2: {monto_venta2:.2f}")

# Intento de venta con cantidad negativa/cero
print('\n--- Intento de venta con cantidad cero ---')
monto_venta3 = inventario.vender('Naranjas', 0)
print(f"Monto de la venta 3: {monto_venta3:.2f}")

print('\n--- Intento de venta con cantidad negativa ---')
monto_venta4 = inventario.vender('Naranjas', -5)
print(f"Monto de la venta 4: {monto_venta4:.2f}")

# Intento de venta con stock insuficiente
print('\n--- Intento de venta con stock insuficiente de Naranjas ---')
monto_venta5 = inventario.vender('Naranjas', 60)
print(f"Monto de la venta 5: {monto_venta5:.2f}")
print(f"Stock actual de Naranjas: {inventario.productos['Naranjas'].stock}")
```

### Verificando el estado final del inventario

```python
print('\n--- Estado final del inventario ---')
print(f"Stock final de Manzanas: {inventario.productos['Manzanas'].stock}")
print(f"Stock final de Naranjas: {inventario.productos['Naranjas'].stock}")
print(f"Stock final de Peras: {inventario.productos['Peras'].stock}")
print(f"Valor total final del inventario: {inventario.obtener_valor_total():.2f}")
```
